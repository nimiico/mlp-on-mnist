```python

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam, SGD, RMSprop
from tensorflow.keras.callbacks import EarlyStopping



```

```python
# Load MNIST data
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Normalize the images
train_images = train_images / 255.0
test_images = test_images / 255.0

# Convert labels to one-hot encoding
train_labels = tf.keras.utils.to_categorical(train_labels, 10)
test_labels = tf.keras.utils.to_categorical(test_labels, 10)
```

```python
#layers

def create_model(num_layers):
    model = Sequential()
    model.add(Flatten(input_shape=(28, 28)))
    for _ in range(num_layers):
        model.add(Dense(128, activation='relu'))
    model.add(Dense(10, activation='softmax'))
    return model

learning_rate = 0.0001
epochs = 40
batch_size = 128

histories = []

for num_layers in range(1, 6):
    model = create_model(num_layers)
    model.compile(optimizer=Adam(learning_rate=learning_rate),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    
    history = model.fit(train_images, train_labels, epochs=epochs, batch_size=batch_size,
                        validation_data=(test_images, test_labels), verbose=2)
    histories.append(history)

# Plotting the accuracy results
plt.figure(figsize=(14, 7))

for i, history in enumerate(histories):
    plt.plot(history.history['val_accuracy'], label=f'{i+1} Layer(s)')

plt.title('Comparison of Model Accuracy with Different Number of Layers')
plt.xlabel('Epoch')
plt.ylabel('Validation Accuracy')
plt.legend()
plt.grid(True)
plt.show()

# Plotting the loss results
plt.figure(figsize=(14, 7))

for i, history in enumerate(histories):
    plt.plot(history.history['val_loss'], label=f'{i+1} Layer(s)')

plt.title('Comparison of Model Loss with Different Number of Layers')
plt.xlabel('Epoch')
plt.ylabel('Validation Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#neurons

def create_model(neurons):
    model = Sequential()
    model.add(Flatten(input_shape=(28, 28)))
    for neuron_count in neurons:
        model.add(Dense(neuron_count, activation='relu'))
    model.add(Dense(10, activation='softmax'))
    return model

learning_rate = 0.0001
epochs = 40
batch_size = 128

neurons_per_layer = [(64, 64), (128, 128), (256, 256), (512, 512), (1024, 1024), (2048, 2048)]

histories = []

for neurons in neurons_per_layer:
    model = create_model(neurons)
    model.compile(optimizer=Adam(learning_rate=learning_rate),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    
    history = model.fit(train_images, train_labels, epochs=epochs, batch_size=batch_size,
                        validation_data=(test_images, test_labels), verbose=2)
    histories.append(history)

# Plotting the loss results
plt.figure(figsize=(14, 7))

for i, history in enumerate(histories):
    plt.plot(history.history['val_loss'], label=f'{neurons_per_layer[i]} Neurons')

plt.title('Comparison of Model Loss with Different Number of Neurons per Layer')
plt.xlabel('Epoch')
plt.ylabel('Validation Loss')
plt.legend()
plt.grid(True)
plt.show()

# Plotting the accuracy results
plt.figure(figsize=(14, 7))

for i, history in enumerate(histories):
    plt.plot(history.history['val_accuracy'], label=f'{neurons_per_layer[i]} Neurons')

plt.title('Comparison of Model Accuracy with Different Number of Neurons per Layer')
plt.xlabel('Epoch')
plt.ylabel('Validation Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#optimizer

def create_model():
    model = Sequential()
    model.add(Flatten(input_shape=(28, 28)))
    model.add(Dense(1024, activation='relu'))
    model.add(Dense(1024, activation='relu'))
    model.add(Dense(10, activation='softmax'))
    return model

learning_rate = 0.0001
epochs = 40
batch_size = 128

optimizers = [Adam(learning_rate=learning_rate), SGD(learning_rate=learning_rate), RMSprop(learning_rate=learning_rate)]
optimizer_names = ['Adam', 'SGD', 'RMSprop', 'Adamax']

histories = []

for optimizer in optimizers:
    model = create_model()
    model.compile(optimizer=optimizer,
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    
    history = model.fit(train_images, train_labels, epochs=epochs, batch_size=batch_size,
                        validation_data=(test_images, test_labels), verbose=2)
    histories.append(history)

# Plotting the loss results
plt.figure(figsize=(14, 7))

for i, history in enumerate(histories):
    plt.plot(history.history['val_loss'], label=f'{optimizer_names[i]} Optimizer')

plt.title('Comparison of Model Loss with Different Optimizers')
plt.xlabel('Epoch')
plt.ylabel('Validation Loss')
plt.legend()
plt.grid(True)
plt.show()

# Plotting the accuracy results
plt.figure(figsize=(14, 7))

for i, history in enumerate(histories):
    plt.plot(history.history['val_accuracy'], label=f'{optimizer_names[i]} Optimizer')

plt.title('Comparison of Model Accuracy with Different Optimizers')
plt.xlabel('Epoch')
plt.ylabel('Validation Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#learning rate

def create_model():
    model = Sequential()
    model.add(Flatten(input_shape=(28, 28)))
    model.add(Dense(1024, activation='relu'))
    model.add(Dense(1024, activation='relu'))
    model.add(Dense(10, activation='softmax'))
    return model

epochs = 40
batch_size = 128

learning_rates = [0.001, 0.0001, 0.00001]
lr_names = ['0.001', '0.0001', '0.00001']

histories = []

for lr in learning_rates:
    model = create_model()
    model.compile(optimizer=Adam(learning_rate=lr),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    
    history = model.fit(train_images, train_labels, epochs=epochs, batch_size=batch_size,
                        validation_data=(test_images, test_labels), verbose=2)
    histories.append(history)

# Plotting the loss results
plt.figure(figsize=(14, 7))

for i, history in enumerate(histories):
    plt.plot(history.history['val_loss'], label=f'Learning Rate: {lr_names[i]}')

plt.title('Comparison of Model Loss with Different Learning Rates (Adam)')
plt.xlabel('Epoch')
plt.ylabel('Validation Loss')
plt.legend()
plt.grid(True)
plt.show()

# Plotting the accuracy results
plt.figure(figsize=(14, 7))

for i, history in enumerate(histories):
    plt.plot(history.history['val_accuracy'], label=f'Learning Rate: {lr_names[i]}')

plt.title('Comparison of Model Accuracy with Different Learning Rates (Adam)')
plt.xlabel('Epoch')
plt.ylabel('Validation Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#early stopping

def create_model():
    model = Sequential()
    model.add(Flatten(input_shape=(28, 28)))
    model.add(Dense(1024, activation='relu'))
    model.add(Dense(1024, activation='relu'))
    model.add(Dense(10, activation='softmax'))
    return model

learning_rate = 0.0001
epochs = 40  # Set to a high value because we will use early stopping
batch_size = 128

# Create the model
model = create_model()
model.compile(optimizer=Adam(learning_rate=learning_rate),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Set up early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train the model with early stopping
history = model.fit(train_images, train_labels, epochs=epochs, batch_size=batch_size,
                    validation_data=(test_images, test_labels), callbacks=[early_stopping], verbose=2)

# Plotting the loss results
plt.figure(figsize=(14, 7))
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.plot(history.history['loss'], label='Training Loss')
plt.title('Model Loss with Learning Rate 0.001 and Early Stopping')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

# Plotting the accuracy results
plt.figure(figsize=(14, 7))
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.title('Model Accuracy with Learning Rate 0.001 and Early Stopping')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#activation function

def create_model(activation):
    model = Sequential()
    model.add(Flatten(input_shape=(28, 28)))
    model.add(Dense(1024, activation=activation))
    model.add(Dense(1024, activation=activation))
    model.add(Dense(10, activation='softmax'))
    return model

learning_rate = 0.0001
epochs = 40  # Set to a high value because we will use early stopping
batch_size = 128

# Activation functions to be tested
activation_functions = ['relu', 'tanh', 'sigmoid', 'elu', 'selu']

histories = {}

for activation in activation_functions:
    model = create_model(activation)
    model.compile(optimizer=Adam(learning_rate=learning_rate),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    # Set up early stopping
    early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

    # Train the model with early stopping
    history = model.fit(train_images, train_labels, epochs=epochs, batch_size=batch_size,
                        validation_data=(test_images, test_labels), callbacks=[early_stopping], verbose=2)
    
    histories[activation] = history

# Plotting the loss results
plt.figure(figsize=(14, 7))

for activation in activation_functions:
    plt.plot(histories[activation].history['val_loss'], label=f'Activation: {activation}')

plt.title('Comparison of Model Loss with Different Activation Functions')
plt.xlabel('Epoch')
plt.ylabel('Validation Loss')
plt.legend()
plt.grid(True)
plt.show()

# Plotting the accuracy results
plt.figure(figsize=(14, 7))

for activation in activation_functions:
    plt.plot(histories[activation].history['val_accuracy'], label=f'Activation: {activation}')

plt.title('Comparison of Model Accuracy with Different Activation Functions')
plt.xlabel('Epoch')
plt.ylabel('Validation Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#dropout

def create_model_with_dropout():
    model = Sequential()
    model.add(Flatten(input_shape=(28, 28)))
    model.add(Dense(1024, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(1024, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(10, activation='softmax'))
    return model

def create_model_without_dropout():
    model = Sequential()
    model.add(Flatten(input_shape=(28, 28)))
    model.add(Dense(1024, activation='relu'))
    model.add(Dense(1024, activation='relu'))
    model.add(Dense(10, activation='softmax'))
    return model

learning_rate = 0.0001
epochs = 40
batch_size = 128

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Model with Dropout
model_with_dropout = create_model_with_dropout()
model_with_dropout.compile(optimizer=Adam(learning_rate=learning_rate),
                           loss='categorical_crossentropy',
                           metrics=['accuracy'])

history_with_dropout = model_with_dropout.fit(train_images, train_labels, epochs=epochs, batch_size=batch_size,
                                              validation_data=(test_images, test_labels),
                                              callbacks=[early_stopping], verbose=2)

# Model without Dropout
model_without_dropout = create_model_without_dropout()
model_without_dropout.compile(optimizer=Adam(learning_rate=learning_rate),
                              loss='categorical_crossentropy',
                              metrics=['accuracy'])

history_without_dropout = model_without_dropout.fit(train_images, train_labels, epochs=epochs, batch_size=batch_size,
                                                    validation_data=(test_images, test_labels),
                                                    callbacks=[early_stopping], verbose=2)

# Plotting the results
plt.figure(figsize=(14, 7))

# Plotting Validation Loss
plt.subplot(1, 2, 1)
plt.plot(history_with_dropout.history['val_loss'], label='With Dropout')
plt.plot(history_without_dropout.history['val_loss'], label='Without Dropout')
plt.title('Comparison of Validation Loss with and without Dropout')
plt.xlabel('Epoch')
plt.ylabel('Validation Loss')
plt.legend()
plt.grid(True)

# Plotting Validation Accuracy
plt.subplot(1, 2, 2)
plt.plot(history_with_dropout.history['val_accuracy'], label='With Dropout')
plt.plot(history_without_dropout.history['val_accuracy'], label='Without Dropout')
plt.title('Comparison of Validation Accuracy with and without Dropout')
plt.xlabel('Epoch')
plt.ylabel('Validation Accuracy')
plt.legend()
plt.grid(True)

plt.show()
```

```python
#batch normalization 

# Define the model with Dropout and Batch Normalization
def create_model():
    model = Sequential()
    model.add(Flatten(input_shape=(28, 28)))
    model.add(Dense(1024, activation='relu'))
    model.add(BatchNormalization())  # Add Batch Normalization
    model.add(Dropout(0.5))  # Add Dropout with rate 0.5
    
    model.add(Dense(1024, activation='relu'))
    model.add(BatchNormalization())  # Add Batch Normalization
    model.add(Dropout(0.5))  # Add Dropout with rate 0.5
    
    model.add(Dense(10, activation='softmax'))
    return model

# Hyperparameters
learning_rate = 0.0001
epochs = 40
batch_size = 128
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Create the model
model = create_model()
model.compile(optimizer=Adam(learning_rate=learning_rate),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
history = model.fit(train_images, train_labels, epochs=epochs, batch_size=batch_size,
                    validation_data=(test_images, test_labels), callbacks=[early_stopping], verbose=2)

# Plotting the results
plt.figure(figsize=(14, 7))

# Plot training & validation accuracy values
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy with Dropout and Batch Normalization')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

# Plot training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss with Dropout and Batch Normalization')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
